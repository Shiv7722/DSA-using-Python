This notebook demonstrates two implementations of a 2D point data structure:
## **Cartesian Coordinates Implementation**  
    - The `Point` class stores points using `x` and `y` coordinates.
    - Methods include translation, addition, and computing distance from the origin.
    - Example usage:
      - Create a point: `point1 = Point(5, 8)`
      - Translate: `point1.translate(0, -4)`
      - Add another point: `point1.__add__(point2)`
      - Compute distance from a point if it is given origin otherwise(Method Overloading): `dist = point1.distorigin()`





```python
import math
```

```python
class Point:
    def __init__(self,x=0,y=0):
        self.x=x
        self.y=y

    def translate(self,del_x,del_y):
        self.x+=del_x
        self.y+=del_y
     
    def __str__(self):
        return f"({self.x}, {self.y})"
    
    def __repr__(self):
        return f"({self.x}, {self.y})"
    
    def __add__(self,point):
        self.x+=point.x
        self.y+=point.y
        return Point(self.x,self.y)
    
    def dist(self,point=None):
        if(point==None):
            return round(math.sqrt((self.x)**2+(self.y)**2),2)
        else:
            return round(math.sqrt((self.x-point.x)**2+(self.y-point.y)**2),2)
```

## Polar Coordinates Implementation

This section demonstrates an alternative `Point` class that internally stores points using polar coordinates: **angle** (from the x-axis) and **distance from the origin**.

### Key Features:
- **Internal Representation:**  
    Each point is stored as an angle and a distance, rather than Cartesian `x` and `y`.
- **Conversion:**  
    The class provides methods to convert between polar and Cartesian coordinates for display and computation.
- **Translation:**  
    The `translate` method shifts the point by a given `(del_x, del_y)` in Cartesian space, updating the internal polar representation.
- **Distance Calculation:**  
    The `length` method computes the distance from the origin or between two points using their polar coordinates.
- **Angle Calculation:**  
    The `anglebw` method returns the angle from the x-axis or the angle between two points.

### below code block define the python implementation

```python
class Point:
    """Point (x, y) """
    def __init__(self, x=0, y=0):
        # if x==0:
        #     self.angle=math.pi/2
        # else:            
        self.angle = math.atan2(y, x)
        self.dist = math.sqrt(x**2 + y**2)
        
    def __str__(self):
        return f"({self.angle}, {self.dist})"
    
    def __repr__(self):
        x=round(math.cos(self.angle)*self.dist,2)
        y=round(math.sin(self.angle)*self.dist,2)
        return f"({x}, {y})"
    
    def translate(self,del_x,del_y):
        x=round(math.cos(self.angle)*self.dist,2)+del_x
        y=round(math.sin(self.angle)*self.dist,2)+del_y
        self.angle=math.atan2(y, x)
        self.dist = math.sqrt(x**2 + y**2)
    
    def length(self, point=None):
        if(point==None):
            return round(self.dist,2)
        else:
            productofdist=self.dist*point.dist
            sumofdistsqr=self.dist**2+point.dist**2
            cosofanglediff=math.cos(self.angle-point.angle)
            return math.sqrt(sumofdistsqr-2*productofdist*cosofanglediff)
    
    def anglebw(self,point=None):
        if(point==None):
            return round(self.angle,2)
        else:
            return round(abs(self.angle-point.angle),2)

    
```

```python
polar = Point(4,6)
polar2 = Point(0,5)
polar2.translate(3,0)

print(polar,polar.length(), polar.anglebw())
print(polar.anglebw(polar2))
polar2
```

2. **Polar Coordinates Implementation**  
    - The `Point` class is redefined to store points using angle (from x-axis) and distance from origin.
    - Internally, it converts polar coordinates to Cartesian (`x`, `y`) for storage.
    - Example usage:
      - Create a point: `point = Point(math.atan(3/4), 5)`



```python
class Point:
    def __init__(self, angle, dist):
        self.x=round(dist*math.cos(angle),2)
        self.y=round(dist*math.sin(angle),2)
        
    def __str__(self):
        return f"({self.x}, {self.y})"
        
    def __add__(self,point):
        x=self.x+point.x
        y=self.y+point.y
        angle=math.atan(y/x)
        dist=math.sqrt(y**2+x**2)
        return Point(angle, dist)
```

```python
point1=Point(math.atan(3/4),5)
point2=Point(math.pi/4,math.sqrt(50))
print(point1,point2)
print(point1.__add__(point2))
```

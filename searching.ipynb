## Searching through a list to check if a element exists in the list or not
Basic or niave or brute force algorithm is to iterate over the whole list untill we find the element matching to the element we are searching for

```python
def exists(l,m):
    for i in l:
        if(i==m):
            return True
    return False
```

```python
l=[4,9,7,12,32,56,76,55,78,98,65,43,56,77]
exists(l,33)
```

## Binary search algorithm
We can optimise this sorting if we have the given list in sorted order

```python
def search(l,m):
    left=0
    right=len(l)-1
    
    while left<=right:
        mid=(left+right)//2
        if(l[mid]==m):
            return True
        elif(l[mid]<m):
            left=mid+1
        elif(l[mid]>m):
            right=mid-1
    return False

```

```python
l=[4,9,7,12,32,56,76,55,78,98,65,43,56,77]
l.sort()
print(l)
search(l,43)
```

## Naive approch to sort a python list


- Below approch create an new list and keeps the minimum appending to the new list 

```python
def insort(l):
    sorted=[]

    while len(l)>0:
        min=l[0]
        min_pos=0
        """below loop finds the minimum and the position 
        of the minimum among the (i+1)th to all the remaining element"""
        for j in range(1,len(l)):
            if l[j]< min:
                min=l[j]
                min_pos=j
        sorted.append(min)
        l.pop(min_pos)
    return sorted

l=[4,17,12,1,7,5,9,14]
sorted=insort(l)
print(sorted)
```

- Inplace sorting without occuping extra memory for an extra list

```python
def insort(l):
    n = len(l)

    for i in range(n):
        min = l[i]
        min_pos=i
        """below loop finds the minimum and the position 
        of the minimum among the (i+1)th to all the remaining element"""
        for j in range(i+1,n):
            if l[j]<min:
                min=l[j]
                min_pos=j
        (l[min_pos],l[i])=(l[i],min) 
    return l  


l=[4,43,56,17,12,1,14,23,19,34,5,7,5,9,14]
sorted=insort(l)
print(sorted)
```

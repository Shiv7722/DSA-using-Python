```python
def factor(n):
    factors=[1,n]
    for i in range(2,n//2):
        if(n%i==0):
            factors.append(i)
    return factors
```

```python
def prime(n):
    fact_list=factor(n)
    if(len(fact_list)==2):
        return n
    else:
        return -1
```

```python
def get_prime(n):
    prime_list=[]
    for i in range(1,n+1):
        if(prime(i)!=-1):
            prime_list.append(i)
    return prime_list

```

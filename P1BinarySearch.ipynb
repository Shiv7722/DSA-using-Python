# Binary Search

```python
arr = [1,2,5,8,9,10,15,20,25,30]
n=5
def binarySearch(arr,n):
    left=0
    right=len(arr)-1
    while(left<right):
        mid=(left+right)//2
        if(arr[mid]==n):
            return [mid,arr[mid]]
        elif(arr[mid]<n):
             left=mid+1
        elif(arr[mid]>n):
             right=mid-1
    if(left==right):
        return [-1,n]
result = binarySearch(arr, 11)
print(result)
```
